Все выполненные доработки должны быть протестированы.

Реализуйте оставшиеся методы этого класса append(item), index(item),
insert(pos,item), pop(), pop(pos).\
Требования к работе этих методов посмотрите в
лекционной презентации.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class UnorderedList:
    def __init__(self):
        self.head = None
    
    def is_empty(self):
        return self.head is None
    
    def add(self, item):
        node = Node(item)
        node.next = self.head
        self.head = node
    
    def size(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count
    
    def remove(self, item):
        current = self.head
        previous = None
        while current:
            if current.data == item:
                if previous is None:
                    self.head = current.next
                else:
                    previous.next = current.next
                return True
            previous = current
            current = current.next
        return False
    
    def search(self, item):
        current = self.head
        while current:
            if current.data == item:
                return True
            current = current.next
        return False
    
    def append(self, item):
        """append(item) добавляет новый элемент в конец списка."""
        new_node = Node(item)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
        
    def index(self, item):
        """index(item) возвращает позицию элемента в списке."""
        current = self.head
        idx = 0
        while current:
            if current.data == item:
                return idx
            current = current.next
            idx += 1
        raise ValueError("Item not found")        
    
    def insert(self, pos, item):
        """insert(pos,item) вставляет новый элемент в заданную позицию pos списка."""
        node = Node(item)
        if pos == 0:
            node.next = self.head
            self.head = node
            return
        current = self.head
        idx = 0
        prev = None
        while current and idx < pos:
            prev = current
            current = current.next
            idx += 1
        if idx != pos:
            raise IndexError("Index out of range")
        prev.next = node
        node.next = current
        
    def pop(self, pos=None):
        """
        pop() удаляет и возвращает последний элемент списка\n
        pop(pos) удаляет и возвращает элемент из позиции pos
        """
        if self.head is None:
            raise IndexError("Pop from empty list")
        if pos is None:
            current = self.head
            prev = None
            while current.next:
                prev = current
                current = current.next
            if prev:
                prev.next = None
            else:
                self.head = None
            return current.data
        else:
            current = self.head
            prev = None
            idx = 0
            while current and idx < pos:
                prev = current
                current = current.next
                idx += 1
            if current is None:
                raise IndexError("Pop index out of range")
            if prev:
                prev.next = current.next
            else:
                self.head = current.next
            return current.data
        
    def __str__(self):
        """содержимое списка отображается в Python-стиле"""
        result = []
        current = self.head
        while current:
            result.append(repr(current.data))
            current = current.next
        return '[' + ', '.join(result) + ']'

    def slice(self, start, stop):
        """принимает два параметра (start и stop) и
        возвращать копию списка,\n
        начиная с позиции start и заканчивая (но не включая!)
        позицией stop.
        """
        result = UnorderedList()
        current = self.head
        idx = 0
        while current and idx < stop:
            if idx >= start:
                result.append(current.data)
            current = current.next
            idx += 1
        return result


ul = UnorderedList()

ul.append(10)
ul.append(20)
ul.append(30)

print(f"Список: {str(ul)}")

ul.insert(1,15)
print(f"Добавлен '15' в позицию '1': {str(ul)}")

print(f"index(20): {ul.index(20)}")

print(f"pop(): {ul.pop()}")

print(f"pop(1): {ul.pop(1)}")

print(f"Список после 'pop': {str(ul)}")

print(f"slice(0,2) {str(ul.slice(0,2))}")
```

Напишите функцию move_to_front для чтения строк с клавиатуры и сохранения их в
связном списке без повторов. Пользователь последовательно вводит с клавиатуры n строк.
При вводе строки, которая раньше не встречалась, вставьте ее в начало списка. При вводе
уже внесенной строки удалите ее из списка и снова вставьте в начало.
Такая функция реализует стратегию «сдвиг в начало», которая применяется в кешировании,
при сжатии данных и т.д.

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def remove(self, data):
        prev = None
        curr = self.head
        while curr:
            if curr.data == data:
                if prev:
                    prev.next = curr.next
                else:
                    self.head = curr.next
                return
            prev = curr
            curr = curr.next

    def exists(self, data):
        curr = self.head
        while curr:
            if curr.data == data:
                return True
            curr = curr.next
        return False

    def __str__(self):
        items = []
        curr = self.head
        while curr:
            items.append(repr(curr.data))
            curr = curr.next
        return "[" + ", ".join(items) + "]"

def move_to_front():
    n = int(input("Введите количество строк: "))
    lst = LinkedList()
    for _ in range(n):
        s = input()
        if lst.exists(s):
            lst.remove(s)
        lst.prepend(s)
    print(lst)


move_to_front() # Ввод: dog cat dog fire cat


```

Используя реализацию связного списка из задания № 1 (класс UnorderedList), реализуйте
следующие линейные структуры (необходимый набор методов посмотрите в лекционной
презентации):
1. Стек.


```python
class Stack:
    def __init__(self):
        self.items = UnorderedList()
    
    def push(self, item):
        self.items.add(item)
    
    def pop(self):
        if self.items.is_empty():
            raise IndexError("pop from empty stack")
        top = self.items.head.data
        self.items.remove(top)
        return top
    
    def peek(self):
        if self.items.is_empty():
            raise IndexError("peek from empty stack")
        return self.items.head.data
    
    def is_empty(self):
        return self.items.is_empty()
    
    def size(self):
        return self.items.size()
```

2. Очередь.


```python
class Queue:
    def __init__(self):
        self.items = UnorderedList()
    
    def enqueue(self, item):
        node = Node(item)
        if self.items.head is None:
            self.items.head = node
        else:
            current = self.items.head
            while current.next:
                current = current.next
            current.next = node
    
    def dequeue(self):
        if self.items.is_empty():
            raise IndexError("dequeue from empty queue")
        front = self.items.head.data
        self.items.head = self.items.head.next
        return front
    
    def is_empty(self):
        return self.items.is_empty()
    
    def size(self):
        return self.items.size()
```

3. Дек.


```python
class Deque:
    def __init__(self):
        self.items = UnorderedList()
    
    def add_front(self, item):
        self.items.add(item)
    
    def add_rear(self, item):
        node = Node(item)
        if self.items.head is None:
            self.items.head = node
        else:
            current = self.items.head
            while current.next:
                current = current.next
            current.next = node
    
    def remove_front(self):
        if self.items.is_empty():
            raise IndexError("remove_front from empty deque")
        front = self.items.head.data
        self.items.head = self.items.head.next
        return front
    
    def remove_rear(self):
        if self.items.is_empty():
            raise IndexError("remove_rear from empty deque")
        current = self.items.head
        prev = None
        while current.next:
            prev = current
            current = current.next
        if prev is None:
            self.items.head = None
        else:
            prev.next = None
        return current.data
    
    def is_empty(self):
        return self.items.is_empty()
    
    def size(self):
        return self.items.size()
```

Реализуйте класс DoubleList для построения двухсвязных списков, в которых каждый узел
содержит ссылку на узел, предшествующий ему в списке, и ссылку на следующий узел.
Класс должен иметь следующие методы: проверка списка на пустоту, поиск элемента,
размер списка, вставка в начало, вставка в конец, удаление из начала, удаление из конца,
вставка перед указанным узлом, вставка после указанного узла, удаление указанного узла.
Проведите тестирование методов реализованного класса.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoubleList:
    def __init__(self):
        self.head = None
        self.tail = None

    def is_empty(self):
        """провека на пустоту"""
        return self.head is None

    def size(self):
        """размер списка"""
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

    def search(self, item):
        """поиск элемента"""
        current = self.head
        while current:
            if current.data == item:
                return True
            current = current.next
        return False

    def insert_front(self, item):
        """вставка в начало"""
        new_node = Node(item)
        if self.is_empty():
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_end(self, item):
        """вставка в конец"""
        new_node = Node(item)
        if self.is_empty():
            self.head = self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def remove_front(self):
        """удаление из начала"""
        if self.is_empty():
            raise IndexError('remove_front from empty list')
        data = self.head.data
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None
        return data

    def remove_end(self):
        """удаление с конца"""
        if self.is_empty():
            raise IndexError('remove_end from empty list')
        data = self.tail.data
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        return data

    def insert_before(self, target_item, item):
        """вставка перед указанным элементом"""
        if self.is_empty():
            raise ValueError('List is empty')
        current = self.head
        while current:
            if current.data == target_item:
                new_node = Node(item)
                prev_node = current.prev
                new_node.next = current
                new_node.prev = prev_node
                current.prev = new_node
                if prev_node:
                    prev_node.next = new_node
                else:
                    self.head = new_node
                return
            current = current.next
        raise ValueError('Target item not found')

    def insert_after(self, target_item, item):
        """вставка после указанного элемента"""
        if self.is_empty():
            raise ValueError('List is empty')
        current = self.head
        while current:
            if current.data == target_item:
                new_node = Node(item)
                next_node = current.next
                new_node.prev = current
                new_node.next = next_node
                current.next = new_node
                if next_node:
                    next_node.prev = new_node
                else:
                    self.tail = new_node
                return
            current = current.next
        raise ValueError('Target item not found')

    def remove(self, item):
        """удаление указаного узла"""
        if self.is_empty():
            raise ValueError('List is empty')
        current = self.head
        while current:
            if current.data == item:
                prev_node = current.prev
                next_node = current.next
                if prev_node:
                    prev_node.next = next_node
                else:
                    self.head = next_node
                if next_node:
                    next_node.prev = prev_node
                else:
                    self.tail = prev_node
                return
            current = current.next
        raise ValueError('Item not found')

    def __str__(self):
        result = []
        current = self.head
        while current:
            result.append(repr(current.data))
            current = current.next
        return '[' + ', '.join(result) + ']'


dl = DoubleList()
print(f"Проверка на пустоту: {dl.is_empty()}")

dl.insert_front(10)
dl.insert_end(20)
dl.insert_front(5)
print(f"Добалено в начало '10', в конец '20', в начало '5'{dl}")

print(f"Размер: {dl.size()}")

print(f"Поиск '10': {dl.search(10)}")

dl.insert_before(10, 7)
print(f"Вставка '7' перед '10': {dl}")

dl.insert_after(10, 15)
print(f"Вставка '15' после '10': {dl}")

dl.remove(7)
print(f"Удаление '7': {dl}")

print(f"Удаление с начала: {dl.remove_front()}")
print(f"Удаление с конца: {dl.remove_end()}")
print(f"Итоговый список: {dl}")
```

Напишите программу, которая, используя структуру данных стек, проверит HTML
документ на соответствие открывающих и закрывающих тегов.
За основу решения можно взять алгоритм проверки скобочной последовательности из
лекционных материалов.

```python
def is_balanced_html(html):
    stack = []
    i = 0
    while i < len(html):
        if html[i] == '<':
            j = html.find('>', i + 1)
            if j == -1:
                return False
            tag = html[i+1:j].strip()
            
            if not tag:
                return False
            
            if tag.startswith('/'):
                tag_name = tag[1:]
                if not stack or stack[-1] != tag_name:
                    return False 
                stack.pop()
            else:
                tag_name = tag.split()[0]
                stack.append(tag_name)
            i = j
        i += 1
    return len(stack) == 0

html1 = """
<html>
  <head>
    <title>Example</title>
  </head>
  <body>
    Hello, world
  </body>
</html>
"""
html2 = "<html><body><p>Test</body></html>"

print(is_balanced_html(html1))
print(is_balanced_html(html2))
```

Стековый калькулятор - это устройство (реальное или виртуальное), которое вычисляет
значения арифметических выражений, записанных в постфиксной форме (также она носит
название Обратная польская запись — Википедия (wikipedia.org)). Стековым он называется
так как действует по следующим правилам: если нам встречается число - мы должны
поместить его в стек; если знак операции - мы должны достать из стека столько чисел,
сколько необходимо для выполнения данной операции и выполнить ее. \
Именно так и работает т.н. математический сопроцессор (например, в INtel8086+AMD64)-
часть ЦП, отвечающая за операции над числами с плавающей запятой. \
Реализуйте стековый калькулятор. Напишите программу, которая читает выражение в
обратной польской нотации и выводит получившееся значение или пишет, что выражение
составлено не корректно (если оно некорректно).

```python
def stack_calculator(expression):
    stack = []
    tokens = expression.split()
    operators = {'+', '-', '*', '/'}
    
    for token in tokens:
        if token not in operators:
            try:
                num = float(token)
                stack.append(num)
            except ValueError:
                return "Ошибка: недопустимый символ"
        else:
            if len(stack) < 2:
                return "Ошибка: недостаточно операндов для операции"
            b = stack.pop()
            a = stack.pop()
            if token == '+':
                stack.append(a + b)
            elif token == '-':
                stack.append(a - b)
            elif token == '*':
                stack.append(a * b)
            elif token == '/':
                if b == 0:
                    return "Ошибка: деление на ноль"
                stack.append(a / b)
    
    if len(stack) != 1:
        return "Ошибка: некорректное выражение"
    return stack[0]

expr1 = "3 4 + 2 * 7 /"
expr2 = "5 1 2 + 4 * + 3 -"
expr3 = "3 4 + +"
expr4 = "4 0 /"
expr5 = "4 ( 1 /"

print(stack_calculator(expr1))
print(stack_calculator(expr2))
print(stack_calculator(expr3))
print(stack_calculator(expr4))
print(stack_calculator(expr5))

```

Используя структуру данных очередь и пример из лекционных материалов, создайте
симуляцию ситуации из реальной жизни \
Вариант 2. Покупатели в очереди на кассе бакалейного магазина. \
Обязательно укажите сделанные вами предположения и предоставьте любые
вероятностные данные, которые нужно будет рассмотреть в качестве части сценария.

```python
import random

class Queue:
    def __init__(self):
        self.items = []
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        if self.is_empty():
            return None
        return self.items.pop(0)
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)

class Customer:
    def __init__(self, id, num_items):
        self.id = id
        self.num_items = num_items
        self.items_left = num_items

def simulate_store():
    queue = Queue()
    time = 0
    customer_id = 1
    current_customer = None
    served_customers = 0

    while time < 50:
        # Приход покупателя
        if random.random() < 0.7:
            items = random.randint(1, 10)
            customer = Customer(customer_id, items)
            queue.enqueue(customer)
            print(f"Время {time}: Новый покупатель {customer_id} с {items} товарами пришёл в очередь")
            customer_id += 1
        
        # Обслуживание текущего покупателя
        if current_customer is None and not queue.is_empty():
            current_customer = queue.dequeue()
            print(f"Время {time}: Покупатель {current_customer.id} начал обслуживание")
        
        if current_customer:
            current_customer.items_left -= 1
            if current_customer.items_left == 0:
                print(f"Время {time}: Покупатель {current_customer.id} обслужен и ушёл")
                served_customers += 1
                current_customer = None
        
        time += 1

    print(f"\nВсего обслужено покупателей: {served_customers}")
    print(f"Осталось в очереди: {queue.size()}")

simulate_store()
```

Предположено, что покупатели приходят с вероятностью (70%) и закупаются случайным количеством товаров (от 1 до 10(вкл)), время обслуживания пропорционально количеству товаров (1 к 1).
