```python
from typing import List, Tuple
import time
import random
import matplotlib.pyplot as plt
import numpy as np

def selection_sort(some_list: List[int]) -> Tuple[List[int], int, int]:
    arr = some_list.copy()
    n = len(arr)
    comparisons = 0
    swaps = 0
    
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
    return arr, comparisons, swaps
```

```python
def print_selection_sort_iterations(some_list: List[int]):
    arr = some_list.copy()
    n = len(arr)
    comparisons = 0
    swaps = 0
    
    print(f" : {arr}\n")
    
    for i in range(n - 1):
        min_idx = i
        iteration_swaps = 0
        iteration_comparisons = 0
        
        for j in range(i + 1, n):
            iteration_comparisons += 1
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
            iteration_swaps = 1
        
        print(f" {i+1}: {arr} (: {iteration_comparisons}, : {iteration_swaps})")
    
    print(f"\n : {arr}")
    print(f" : {comparisons}")
    print(f" : {swaps}")
    return comparisons, swaps

print(" 1\n")
comp1, swaps1 = print_selection_sort_iterations([7, 3, 9, 4, 2, 5, 6, 1, 8])
```

```python

def generate_random_list(n: int):
    """  """
    return [random.randint(1, 10000) for _ in range(n)]

def generate_reverse_sorted_list(n: int):
    """    """
    return list(range(n, 0, -1))

def generate_almost_sorted_list(n: int):
    """  """
    arr = list(range(1, n + 1))
    if n > 1:
        mid = n // 2
        arr[0], arr[mid] = arr[mid], arr[0]
    return arr

def generate_half_zeros_list(n: int):
    """  - 0,  - 
      """
    arr = [0] * (n // 2) + [random.randint(1, 10000) for _ in range(n - n // 2)]
    random.shuffle(arr)
    return arr

def generate_close_to_position_list(n: int):
    """     10- 
        """
    arr = list(range(1, n + 1))
    for i in range(n):
        offset = random.randint(-10, 10)
        new_pos = max(0, min(n - 1, i + offset))
        if i != new_pos:
            arr[i], arr[new_pos] = arr[new_pos], arr[i]
    return arr

```

```python
sizes = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000]

generators = {
    'random': (generate_random_list, 'Random Numbers'),
    'reverse_sorted': (generate_reverse_sorted_list, 'Reverse Sorted'),
    'almost_sorted': (generate_almost_sorted_list, 'Almost Sorted'),
    'half_zeros': (generate_half_zeros_list, 'Half Zeros'),
    'close_position': (generate_close_to_position_list, 'Close to Position')
}

results = {}

print(" 4")

for gen_key, (gen_func, gen_label) in generators.items():
    results[gen_key] = {'times': [], 'sizes': sizes}
    print(f"\n : {gen_label}")
    print("-" * 80)
    
    for size in sizes:
        data = gen_func(size)
        start = time.perf_counter()
        selection_sort(data.copy())
        elapsed = time.perf_counter() - start
        results[gen_key]['times'].append(elapsed)
        print(f"  n={size:5d}: {elapsed:10.6f} ")

print("\n.")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Selection Sort Performance on Different Data Types', fontsize=16, fontweight='bold')
axes = axes.flatten()

for idx, (gen_key, data) in enumerate(results.items()):
    if idx < 5:
        ax = axes[idx]
        ax.plot(data['sizes'], data['times'], 'o-', linewidth=2, markersize=8)
        ax.set_xlabel('Input Size (n)', fontsize=11, fontweight='bold')
        ax.set_ylabel('Time (seconds)', fontsize=11, fontweight='bold')
        ax.set_title(list(generators.values())[idx][1], fontsize=12, fontweight='bold')
        ax.grid(True, alpha=0.3)

axes[5].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
def selection_sort_dual(some_list: List[int]) -> Tuple[List[int], int, int]:
    arr = some_list.copy()
    n = len(arr)
    comparisons = 0
    swaps = 0
    left = 0
    right = n - 1
    
    while left < right:
        min_idx = left
        max_idx = right
        
        for i in range(left, right + 1):
            comparisons += 1
            if arr[i] < arr[min_idx]:
                min_idx = i
            comparisons += 1
            if arr[i] > arr[max_idx]:
                max_idx = i
        
        if min_idx != left:
            arr[left], arr[min_idx] = arr[min_idx], arr[left]
            swaps += 1
            if max_idx == left:
                max_idx = min_idx
        
        if max_idx != right:
            arr[right], arr[max_idx] = arr[max_idx], arr[right]
            swaps += 1
        
        left += 1
        right -= 1
    
    return arr, comparisons, swaps
```

```python
results_dual = {}
print(" 5")

for gen_key, (gen_func, gen_label) in generators.items():
    results_dual[gen_key] = {'times': [], 'sizes': sizes}
    print(f"\n : {gen_label}")
    
    for size in sizes:
        data = gen_func(size)
        start = time.perf_counter()
        selection_sort_dual(data.copy())
        elapsed = time.perf_counter() - start
        results_dual[gen_key]['times'].append(elapsed)
        print(f"  n={size:5d}: {elapsed:10.6f} ")

print("\n")
```

```python
print("\n  ( n=10000)\n")
print(f"{' ':<20} | {'Selection Sort':<15} | {'Selection Sort Dual':<15} | {'':<12}")
print("-" * 80)

for gen_key, data in results.items():
    ss_time = data['times'][-1]
    dual_time = results_dual[gen_key]['times'][-1]
    improvement = (ss_time - dual_time) / ss_time * 100
    speedup = ss_time / dual_time
    
    gen_label = list(generators.values())[[*generators.keys()].index(gen_key)][1]
    print(f"{gen_label:<20} | {ss_time:>14.6f}s | {dual_time:>18.6f}s | {improvement:>10.1f}%")
    print(f"{'':<20} | {'':>15} | {speedup:>18.2f}x |")
    print("-" * 80)
```
